```python
means = df_events.mean()
standard_errors = df_events.sem()
```

```python
# Define the thresholds
lower_bound = means - 2 * standard_errors
upper_bound = means + 2 * standard_errors
```

```python
def vectorized_categorize(values, lower, upper):
    # Initialize result array with '+/-'
    result = np.full_like(values, '', dtype=str)
    # Find indices where values are below lower bound
    below_lower_indices = values < lower
    # Find indices where values are above upper bound
    above_upper_indices = values > upper
    # Assign '-' to values below lower bound
    result[below_lower_indices] = '-'
    # Assign '+' to values above upper bound
    result[above_upper_indices] = '+'
    return result
```

```python
# Apply the vectorized function to the DataFrame
categorized_values = vectorized_categorize(df_events.values, lower_bound.values, upper_bound.values)
```

```python
# Create a new DataFrame from the categorized values
categorized_df = pd.DataFrame(categorized_values, columns=df_events.columns, index=df_events.index)
```

```python
# Specify the path to your .xlsx file
rules_path = 'Cells markers - v2.xlsx'

# Read the Excel file into a DataFrame
rules_df = pd.read_excel(rules_path)
```

```python
# Merge the first four columns into one
rules_df['cell_type'] = [''.join(row) for row in zip(rules_df['Unnamed: 0'].fillna(''),
                                                   rules_df['Unnamed: 1'].fillna(''),
                                                   rules_df['Unnamed: 2'].fillna(''),
                                                   rules_df['Unnamed: 3'].fillna(''))]

# Drop the individual columns after merging
rules_df = rules_df.iloc[:, 4:]
rules_df
```

```python
categorized_df = categorized_df.head(25)
categorized_df
```

```python
# Function to match a row with the rule dataframe
def match_cell_type(row, rule_df):
    try:
        matches = []
        for _, rule in rule_df.iterrows():
            match = True
            for col in rule_df.columns[1:-1]:  # Exclude the 'cell_type' column
                if type(rule[col]) == str and row[col] != rule[col]:
                    match = False
                    break
            if match:
                matches.append(rule['cell_type'])

        return matches if matches else ['Unknown']
    except:
        return ['Unknown']
```

```python
categorized_df['cell_type'] = categorized_df.apply(match_cell_type, axis=1, rule_df=rules_df)
```

```python
import matplotlib.pyplot as plt

# Counting rows with 'Live immune cells'
live_count = categorized_df['cell_type'].apply(lambda x: 'Lymphocytes' in x).sum()

# Counting other rows
other_count = len(categorized_df) - live_count

# Creating a pie chart
labels = ['Live immune cells', 'Other']
sizes = [live_count, other_count]
colors = ['lightblue', 'lightcoral']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of cell types')
plt.show()
```
